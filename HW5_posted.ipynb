### Problem 4
In this problem, we want to develop a classifier using the MLP implementation of Scikit-Learn (https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html). We start by creating a synthetic data set. 

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

n_samples = 200
blob_centers = ([1, 1], [3, 4], [1, 3.3], [3.5, 1.8])
data, labels = make_blobs(n_samples=n_samples, 
                          centers=blob_centers, 
                          cluster_std=0.5,
                          random_state=0)


colours = ('green', 'orange', "blue", "magenta")
fig, ax = plt.subplots()

for n_class in range(len(blob_centers)):
    ax.scatter(data[labels==n_class][:, 0], 
               data[labels==n_class][:, 1], 
               c=colours[n_class], 
               s=30, 
               label=str(n_class))
```

(a) Divide the available data into train and test sets with the following parameters: test_size=0.2, random_state=42

```python
# Import Libraries
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neural_network import MLPClassifier
import numpy as np
```

```python
# Split the data into train and test sets (80/20) split
X_train, X_test, y_train, y_test = train_test_split(data, labels, 
                                                    test_size= 0.2, 
                                                    train_size=0.8, 
                                                    random_state=42)
```

(b) Implement and train a neural network with only one hidden layer. Use GridSearchCV to find the optimal number of units in the hidden layer. 

```python
# Develop the MLP Classifier
MLP = MLPClassifier(random_state=42, max_iter=1000)

# GridSearch
param_grid = [{"hidden_layer_sizes": np.linspace(start = 1, stop = 100, 
                                                 num = 100, dtype = int)}]
grid_search = GridSearchCV(MLP, param_grid, scoring='accuracy', cv=5)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```
